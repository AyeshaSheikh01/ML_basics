### Titanic dataset

Today, we will be working on titianic dataset using the github url. 

```python
import pandas as pd 
pd.set_option('display.max_columns', None) # Show all columns
pd.set_option('display.width', 1000) # doesnt wrap the columns
```

<h5> Importing dataset via URL <h5>

```python

url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv?utm_source=chatgpt.com" 
pd.read_csv(url, header=None)
```

```python
# Changing header names  
# header must be overwrite again. 
df=pd.read_csv( 
    url, 
    header=0, 
    names=['id', 'survived', 'passengerClass', 'Name', 'Gender', 'Age', 'SiblingsSpouseNumber', 'ParentsChildNumbers','TicketNumber', 'Fare', 'Cabin', 'Embarked']
    )  
print(df.head(5))
```

```python
# use of index_col and usecols 
# index_cols is to tell panda not to use the first cols as the header, set it False in parameters.  
df1=pd.read_csv(url, index_col='PassengerId', dtype={'Survived':'float'}, usecols=['PassengerId','Survived','Pclass'])  # choosing selective cols 
print(df1.head(5)) #DTYPE CAN CHANGE THE TYPE OF THE VARIABLE
```

```python
# skipping rows 
df2=pd.read_csv(url) 
print(df2.head(5))
```

```python
df3=pd.read_csv(url, skiprows=[0,1,2,3,4], skip_blank_lines=True) 
print(df3) # looked it skipped the rows that I asked it to.
```
