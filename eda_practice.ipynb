### Working on EDA using titanic dataset

```python
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv?utm_source=chatgpt.com"  
titanicDataset=pd.read_csv(url, sep=",")  
print(titanicDataset)
```

<h4> Exploratory data Analysis </h4>

```python
# 1- check the size of the dataset 
titanicDataset.shape 
```

```python
#2- how the data is organized 
titanicDataset.head(5)
```

```python
# 3- finidng the data types and we can change it too using dtype parat=meter of read_csv  
# tells about the memory usage too
titanicDataset.info() 
```

```python
titanicDataset.describe() #only works for numerical columns
```

Let's find out the insights that we can find using this describe() function. If we look at the survival rate, we can derive that it is very low as mean is very low. And for age, avg age was of 30 while max was 80. 

```python
# 4- checking missing values 
titanicDataset.isnull().sum()
```

```python
# 5- Find the duplicate values if exists 
titanicDataset.duplicated() # to check separately
```

```python
# but finding the sum of the duplicate is easier.  
titanicDataset.duplicated().sum()
```

```python
# finding the correlation between cols.  
titanicDataset.corr(numeric_only=True)['Survived']
```

<h4> > use of value_counts() </h4>

```python
titanicDataset["Sex"].value_counts() 
```

```python
titanicDataset["Pclass"].value_counts()
```

```python
titanicDataset["Embarked"].value_counts()
```

Look, it shows what kind of data is in categorical cols. It is good for categories. By using this, we can observe sex has two categories and there ranges (male and female), Pclass has three class plus their comparison. The 3rd class was more in numbers. For embarked, we can see where passenger embraked and mostly were from S.  
C = Cherbourg, Q = Queenstown, S = Southampton

But for numerical data, we tend to go for describe because it might give us insights in mathematical terms. 

```python
titanicDataset["Survived"].describe()
```

<h4> > use of groupby() </h4> 
"Group the dataframe by this column, then perform this operation on another column."

```python
titanicDataset.groupby("Pclass")["Age"].mean()
```

showing passengers in the first class were mostly older than passengers in 1st and 2nd.
